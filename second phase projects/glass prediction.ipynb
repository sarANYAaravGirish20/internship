```python
X = df.drop(columns='Type',axis=1)
y = df['Type']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
from sklearn import metrics
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
# Decision Tree
DeciTree = tree.DecisionTreeClassifier()
DeciTree.fit(X_train, y_train)
y_pred = DeciTree.predict(X_test)
acc = metrics.accuracy_score(y_pred,y_test)
print(acc)
```

```python
# Random Forest
RandFore = RandomForestClassifier(max_depth = 3,random_state = 1)
RandFore.fit(X_train, y_train)
y_pred = RandFore.predict(X_test)
acc1 = metrics.accuracy_score(y_pred,y_test)
print(acc1)
```
