```python
upload=[]
try:
    date=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]//tr//td[5]')
    for i in date:
        upload.append(i.text)
except NoSuchElementException:
    upload.append('-')
except StaleElementReferenceException:
    upload.append('-')
len(upload)
```
