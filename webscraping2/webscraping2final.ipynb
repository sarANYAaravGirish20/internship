Q1:Write a python program to scrape data for “Data Analyst” Job position in“Bangalore” location. You
have to scrape the job-title, job-location, company_name. You have to scrape first 10 jobs data.
This task will be done in following steps:
1. First get the webpage https://www.shine.com/
2. Enter “Data Analyst” in “Job title, Skills” field and enter “Bangalore” in “enter thelocation” field.
3. Then click the search button.
4. Then scrape the data for the first 10 jobs results you get.
5. Finally create a dataframe of the scraped data.
Note: All of the above steps have to be done in code. No step is to be done manually.

```python
driver=webdriver.Chrome()
driver.get("https://www.shine.com")
```

```python
designation=driver.find_element(By.CLASS_NAME,"form-control  ")
designation.send_keys('Data Analyst')
```

```python
location = driver.find_element(By.XPATH,"/html/body/div[1]/div[4]/div/div[2]/div[2]/div/form/div/div[1]/ul/li[2]/div/input")
location.send_keys('Bangalore')
```

```python
search =driver.find_element(By.CLASS_NAME,"searchForm_btnWrap_advance__VYBHN")
search.click()
```

```python

job_title=[]
title_tags=driver.find_elements(By.XPATH,'//h2[@itemprop="name"]')
for i in title_tags[0:10]:
    job_title.append(i.text)
    
company_Name=[]
company_tags=driver.find_elements(By.XPATH,'//div[@class="jobCard_jobCard_cName__mYnow"]')
for i in company_tags[0:10]:
    company_Name.append(i.text)

job_location=[]
location_tags=driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_locationIcon__zrWt2"]')
for i in location_tags[0:10]:
    job_location.append(i.text)
    
Experience=[]
exp_tags=driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_jobIcon__3FB1t"]')
for i in exp_tags[:10]:
    Experience.append(i.text)
```

```python
df=pd.DataFrame({'Job_title':job_title[0:10],'Job_location':job_location[0:10],'Company_name':company_Name[0:10],'Exp_required':Experience[0:10]})
df
```

```python
driver.close()
```

```python
driver=webdriver.Chrome()
driver.get("https://www.amazon.in")
```

```python
lap=driver.find_element(By.XPATH,'//*[@id="twotabsearchtextbox"]')
lap.send_keys('laptop')
```

```python
cpu_filt=driver.find_element(By.XPATH,'/html/body/div[1]/div[2]/div[1]/div[2]/div/div[3]/span/div[1]/div/div/div[7]/ul[7]/span[13]/li/span/a/span')
cpu_filt.click()
```

```python
Laptitle=[]
lapprice=[]
Rate=[]
start=0
end=1
for page in range(start,end):
    title=driver.find_elements(By.XPATH,"//h2[@class='a-size-mini a-spacing-none a-color-base s-line-clamp-2']")
    for i in title[0:10]:
        Laptitle.append(i.text) 
        
    prices=driver.find_elements(By.XPATH,'//span[@class="a-price-whole"]')
    for i in prices[0:10]:
        lapprice.append(i.text)  
```

```python
titles_rat=driver.find_elements(By.XPATH,"//div[@class='a-row a-size-small']/span")
for i in titles_rat:
    Rate.append(i.get_attribute("aria-label"))
rating=[]
for i in range(0,len(Rate))[0:20]:
    if i == 0 or  i/2 == i//2:
        rating.append(Rate[i][0:3])
```

```python
df=pd.DataFrame({'Title':Laptitle,'Price':lapprice,'Rating':rating})
df
```
